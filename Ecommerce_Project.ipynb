```python
# import pandas library 
import pandas as pd 
#import matplotlib library
import matplotlib.pyplot as plt 
```

```python
#load ecommerce_sales_analysis file 
data=pd.read_csv("ecommerce_sales_analysis.csv")
```

```python
#showing first and last five rows of the data 
data.head()
data.tail()
```

```python
#checking whether their are any null values in our data 
data.isnull().sum()
#their are no null values in data 
```

```python
#visualization of data 
#import seaborn for visualization of data 
import seaborn as sns
```

```python
#category of product and its price 
sns.barplot (x='category',y='price',data=data)
plt.show()
```

```python
#Correlation between price, review_score, review_count, and total sales.
sns.heatmap(data[['price', 'review_score', 'review_count']].corr(), annot=True)

```

```python
#Review scores distribution by category.
sns.boxplot(x='category',y='review_score', data=data,color='b')
```

```python
#Price vs Review Score  Are expensive products getting better reviews?
sns.scatterplot(x='price',y='review_score',data=data,color='y')
plt.show()
```

```python
#Number of products in each category.
sns.countplot (x='category',data=data,color='r')
plt.show()
```
